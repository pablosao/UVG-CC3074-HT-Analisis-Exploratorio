```python
import pandas as pd
```

```python
movie = pd.read_csv('tmdb-movies.csv')
```

```python
print (movie.columns)
```

```python
director_movies_revenue = movie.iloc[:,[8,4]]
```

```python
director_movies_revenue.groupby('director')
```

```python
director_movies_revenue
```

```python
directores_clean = director_movies_revenue.loc[director_movies_revenue['revenue'] != 0]
```

```python
directores_clean
```

```python
# se agregar otra columna
directores_clean['count_movies'] = 1
```

```python
directores_clean
```

```python
directores_ready = directores_clean.groupby(['director'], sort = False).sum()
```

```python
directores_ready
```

```python
print (directores_ready.index)
```

```python
import plotly.graph_objects as go

fig = go.Figure(data=[go.Scatter(
    x=directores_ready['count_movies'], y=directores_ready['revenue'],
    text=directores_ready.index,
    mode='markers',
    marker=dict(
        color=['rgb(0, 0, 255)']
    )
)])

fig.show()
```
